```python
import geopandas as gp
uganda_shapes = gp.GeoDataFrame.from_file("/home/nathan/GPM-district-precipitation/District_Boundaries_2014.shp")
def capitalize(s):
    return " ".join(
        map(lambda k: k[0].upper() + k[1:], s.lower().split(' ')))
uganda_shapes['district'] = uganda_shapes.DNAME2014.map(capitalize)
uganda_shapes['area_km'] = uganda_shapes.to_crs('+init=epsg:3395').area / (1000 ** 2)
uganda_shapes.to_pickle('uganda_shapes.p')
uganda_shapes
```

```python
import pandas as pd
gpm = pd.read_pickle('GPM-timeseries-by-district-uganda.p')
gpm
#TODO: Add district coastline property
```

```python
import json
with open('districts.json') as f:
    districts_by_id = json.load(f)
```

```python
import requests
try:
    case_data = pd.read_pickle('case-data.p')
except IOError:
    case_data = pd.DataFrame()
    for district_id, district in districts_by_id.items():
        resp = requests.get('http://50.116.56.97/data.json',
            params=dict(
                from_date="201012",
                to_date="201612",
                region="0",
                district=district_id,
                age_group="0",
                group="period"))
        case_data = case_data.append([
            dict(district=district,
               district_id=district_id,
               week=week,
               **item)
            for week, item in resp.json().items()])
    case_data.to_pickle('case-data.p')
case_data
```

```python
import datetime
import numpy as np
def parse_week(week_str):
    if "W" in week_str:
        [year, week_num] = week_str.split("W")
        return pd.datetime(int(year), 1, 1) +\
            datetime.timedelta(days=int(week_num) * 7)
    else:
        return pd.datetime(int(week_str[:4]), int(week_str[5:]), 1)
d = pd.DataFrame(case_data)
d = d[d.week.str.contains('W')]
d.week = d.week.map(parse_week)
d['district'] = d.district.map(lambda x: x.replace(" District", ""))
gpm_by_week = gpm.groupby(['week', 'district']).sum().reset_index()
d = d.merge(gpm_by_week,
            left_on=["week", "district"],
            right_on=["week", "district"],
            suffixes=('', ''))

d = d.merge(uganda_shapes, left_on=["district"], right_on=["district"])

d['population_density'] = d.population / d.area_km
d['case_ratio'] = d.malaria_cases_wep / d.population
d[['malaria_cases_wep', 'district_mean_rainfall_mm', 'population']].corr()
```

```python
plotd = d[[
    'malaria_cases_wep', 'district_mean_rainfall_mm', 'week', 'district'
]]
plotd['district_mean_rainfall_mm'] *= 100
plotd.query('district == "Lamwo"').set_index('week').plot(figsize=(12, 8))
```

```python
BASELINE_WEEKS = 5
FUTURE_START_END_WEEKS = (8, 12)
def agg_fun(group):
    group = group.sort('week', ascending=False)
    group['mean_future_cases'] = (\
        group.malaria_cases_wep.rolling(FUTURE_START_END_WEEKS[1] + 1).sum() -\
        group.malaria_cases_wep.rolling(FUTURE_START_END_WEEKS[0] + 1).sum()) / (FUTURE_START_END_WEEKS[1] - FUTURE_START_END_WEEKS[0])
    group = group.sort('week')
    # An outbreak is defined as the average number of cases over the future interval
    # exceeding the average number of cases over the past inverval by at least one
    # standard deviation of the past interval.
    group['future_outbreak'] = ((group['mean_future_cases'] -\
        pd.rolling_mean(group['malaria_cases_wep'], BASELINE_WEEKS)) / pd.rolling_std(group['malaria_cases_wep'], BASELINE_WEEKS) > 1)
    group['rainy_days'] = pd.rolling_mean(group.rainy_day, BASELINE_WEEKS)
    group['rainfall_anomoly'] = ((group.district_mean_rainfall_mm -\
        pd.rolling_mean(group.district_mean_rainfall_mm, BASELINE_WEEKS)) / pd.rolling_std(group.district_mean_rainfall_mm, BASELINE_WEEKS) > 1) 
    group['cases_anomoly'] = ((group.malaria_cases_wep - pd.rolling_mean(group.malaria_cases_wep, BASELINE_WEEKS)) /\
        pd.rolling_std(group.malaria_cases_wep, BASELINE_WEEKS) > 1)
    group['mean_future_case_multiple'] = group.mean_future_cases / pd.rolling_mean(group.malaria_cases_wep, BASELINE_WEEKS)
    return group

data = d.groupby(['district']).apply(agg_fun)
data = data[data.week >= (data.week.min() + pd.Timedelta(BASELINE_WEEKS * 7, 'd'))]
data.query('district == "Lamwo"')[[
    'mean_future_cases',
    'week', 'malaria_cases_wep'
]].groupby('week').mean().plot(figsize=(12, 8))
data.query('district == "Lamwo"')[[
    'week', 'future_outbreak', 'district_mean_rainfall_mm',
    'rainfall_anomoly', 'cases_anomoly'
]].groupby('week').mean().plot(figsize=(12, 8))
data[['cases_anomoly', 'district_mean_rainfall_mm', 'future_outbreak', 'population', 'rainy_days', 'rainfall_anomoly']].corr()
```

```python
import sklearn
from sklearn.feature_extraction import DictVectorizer
from sklearn import tree
from IPython.display import Image  
import pydotplus
import numpy as np

vectorizer = DictVectorizer()
data_with_none = data.fillna("none")
dicts = [row[1].to_dict() for row in data_with_none[[
            'cases_anomoly',
            'population_density', 'case_ratio', 'population',
            'district_mean_rainfall_mm', 'rainy_days', 'rainfall_anomoly']].iterrows()]
targetdicts = [row[1][0] for row in data_with_none[['future_outbreak']].iterrows()]
Xt = vectorizer.fit_transform(dicts).toarray()

Xt_binarized = np.array(Xt)
for i in range(Xt.shape[1]):
    med = np.median(Xt[:,i])
    if med == 0:
        Xt_binarized[:,i] = Xt[:,i] > med
    else:
        Xt_binarized[:,i] = Xt[:,i] >= med

def sanitize(s):
    return s.replace("&", "and")

class_names = []
for x in targetdicts:
    x = str(x).replace('&', '')
    if x not in class_names:
        class_names.append(x)

print(class_names.sort())
clf = tree.DecisionTreeClassifier(
    max_leaf_nodes=16,
    min_samples_leaf=.02)
clf = clf.fit(Xt_binarized, targetdicts)
dot_data = tree.export_graphviz(
    clf, out_file=None,
    proportion=True,
    feature_names=[sanitize(n) for n in vectorizer.get_feature_names()],
    class_names=class_names,
    filled=True, rounded=True,
    special_characters=True)
graph = pydotplus.graph_from_dot_data(str(dot_data))
Image(graph.create_png())
```
